## 2. Conceptual Layout (SkyRL-style)

A minimal SkyRL project typically contains:

- **Environment**: emits `observation`, `reward`, `done`
- **Policy / Agent**: maps observation → action
- **Trainer / Loop**: collects rollouts and updates the policy

In real runs, the *policy* might wrap an LLM and the environment provides tasks/rewards.

## 3. Dummy Environment (Safe Placeholder)
A single-step environment to demonstrate the interface. No external dependencies.

```python
class DummyEnv:
    """A one-step environment for demonstration only."""
    def reset(self):
        return {'obs': 'hello'}

    def step(self, action):
        obs = {'obs': 'world'}
        reward = 0.0
        done = True
        info = {'action': action}
        return obs, reward, done, info

env = DummyEnv()
env.reset()
```

## 4. Dummy Agent (Safe Placeholder)
A minimal agent that always returns a constant action.

```python
class DummyAgent:
    """Returns a constant action; replace with an LLM policy wrapper."""
    def act(self, obs):
        return 'noop'

agent = DummyAgent()
obs = env.reset()
action = agent.act(obs)
print('obs:', obs)
print('action:', action)
print('step:', env.step(action))
```

## 5. Rollout Skeleton
A tiny rollout loop showing how environment + agent interact.

```python
def run_one_episode(env, agent):
    traj = []
    obs = env.reset()
    done = False
    while not done:
        action = agent.act(obs)
        next_obs, reward, done, info = env.step(action)
        traj.append({'obs': obs, 'action': action, 'reward': reward, 'next_obs': next_obs, 'info': info})
        obs = next_obs
    return traj

traj = run_one_episode(env, agent)
traj
```
