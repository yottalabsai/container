## 2. Conceptual Layout (Miles-style)

A common Miles-style RL pipeline is split into the following roles:

- **Workers / Rollout**  
  Generate trajectories: `prompt → response → reward`
- **Buffer / Dataset**  
  Store experiences for training
- **Learner**  
  Update policy / value models
- **Evaluator**  
  Periodic evaluation and logging

> ⚠️ In real usage, Miles is usually run via **CLI / scripts** on clusters.
